```python
import re#Regular Expressions module to search, match, or clean text
import nltk#Imports the Natural Language Toolkit library for Natural Language Processing (NLP) tasks.
from nltk.corpus import stopwords#Imports stopwords (common words like is, am, the) used for text cleaning.
from nltk.tokenize import word_tokenize#Imports word tokenizer to split a sentence into individual words.
```

```python
nltk.download('punkt')#punkt is used for tokenization (splitting text into words/sentences).
nltk.download('punkt_tab')
nltk.download('stopwords')#remove common words
nltk.download('wordnet')#Used for lemmatization (convert words to root form).
```

```python
text="Wow!!! The iphone cost in 2026 is $999.99. Can you believe it???"
text=text.lower()
print(text)
clean_text=re.sub(r'[^a-zA-Z\s]','',text)#prints all including spaces without numbers and special characters
print(clean_text)
clean_text=re.sub(r'[^a-zA-Z0-9\s]','',text)#prints all including spaces(\s) without special characters
print(clean_text)
```

```python
words=word_tokenize(clean_text)
print(words)
```

```python
#stopwords
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
#remove stop_words
words=[word for word in words if word not in stop_words]
print(words)
```

```python
#Lemmatizer ->rootword
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = [lemmatizer.lemmatize(word) for word in words]
print(words)
```

BAG OF WORDS

```python
#BAG OF WORDS->Converts text into numbers (BOW)
#convert raw text into a table of numbers
from sklearn.feature_extraction.text import CountVectorizer
Sentences=["I ate a pizza in the morning",
           "I ate chappathi"]
cv=CountVectorizer()#counts words frequence
bow=cv.fit_transform(Sentences)
#finds all unique words in an order
#ate
#chappathi
#in
#morning
#pizza
#the
print(bow.toarray())#This converts output into normal matrix form.
print(cv.get_feature_names_out())
```

TF-IDF (Term Frequency – Inverse Document Frequency)

```python
#TF-IDF (Term Frequency – Inverse Document Frequency)
from sklearn.feature_extraction.text import TfidfVectorizer
Documents=["I am playing cricket",
           "He played cricket yesterday"]
tfidf=TfidfVectorizer()
result=tfidf.fit_transform(Documents)
print(result.toarray())
print(tfidf.get_feature_names_out())
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
texts = [
    'virat kohli scored an amazing century yesterday',
    'hate waiting long traffic jam every morning',
    'openai launched powerful new ai model california',
    'pizza extremely delicious',
    'amazon delivered package late badservice',
    'dhoni love playing cricket young player',
    'microsoft announced new cloud service new york'
]
texts = [sentence.lower() for sentence in texts]

print(texts)
```

```python
clean_texts=[
    re.sub(r'[^a-zA-Z\s]','',sentence)
    for sentence in texts
    ]
print(clean_texts)
```

```python
stop_words = set(stopwords.words('english'))

filtered_texts = [
    [word for word in sentence.split()
     if word not in stop_words]
    for sentence in clean_texts
]

print(filtered_texts)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words =[ [lemmatizer.lemmatize(word) for word in sentence]
        for sentence in filtered_texts
         ]
print(words)
```

Named Entity Recognition(NER)

```python
import spacy
#used for advanced NLP tasks like NER, POS tagging, tokenization, and lemmatization.
```

```python
nlp=spacy.load("en_core_web_sm")
text="Akhil works at Google in Hydrebad"
doc=nlp(text)
for ent in doc.ents:
    print("Entity:",ent.text)
    print("Lable:",ent.label_)
```

Text Classification

```python
from nltk.downloader import StaleMessage
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
text=["AI is growing fast","Google launches new AI","Stockmarket crashed","Bitcoin price increased"]
labels=['Technology','Technology','Finance','Finance']
cv=CountVectorizer()
x=cv.fit_transform(text)
model=MultinomialNB()
model.fit(x,labels)
#model is trained with the given data
```

```python
#Testing the model with an example
test=['Apple released a new phone with ai features']
test_vector=cv.transform(test)
prediction=model.predict(test_vector)
print(prediction)
```

```python
from nltk.downloader import StaleMessage
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
text = [
    "cute kitten meow",
    "small cat drinks milk",
    "cat likes fish",
    "dog barks loudly",
    "puppy plays outside",
    "dog likes bones"
]
labels = [
    "Cat",
    "Cat",
    "Cat",
    "Dog",
    "Dog",
    "Dog"
]
cv=CountVectorizer()
x=cv.fit_transform(text)
model=MultinomialNB()
model.fit(x,labels)
test=['cat plays with ball']
test_vector=cv.transform(test)
prediction=model.predict(test_vector)
print(prediction)
```
