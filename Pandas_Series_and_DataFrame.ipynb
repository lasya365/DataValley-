```python
import pandas as pd
data={
    "Names":['A',None,'B','C'],
    "Age":[None,23,20,None]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.isnull())#it return "True" for missing values,and return "False" for Present values
print(df.isnull().sum())#this counts the missing values
```

```python
#fillna() → Fill missing values
#inplace=True → Save changes in original data directly
df["Age"].fillna(df["Age"].median(),inplace=True)
df["Names"].fillna(df["Names"].mode()[0],inplace=True)
print(df)
#The error is it creates a temporary copy sometimes, and inplace=True tries to modify that copy, not the original DataFrame.so use below way to modify
df["Age"]=df["Age"].fillna(df["Age"].median())
df["Names"]=df["Names"].fillna(df["Names"].mode()[0])
print(df)
```

```python
import pandas as pd
data={
    "Names":['A','B','A','C'],
    "Age":[21,23,21,20]
}
df=pd.DataFrame(data)
print(df)
```

```python
print(df.duplicated())  # finds repeated rows (True/False)
print(df.duplicated().sum())  # counts repeated rows
print(df.drop_duplicates())  # deletes repeated rows (temporary, original df unchanged)
df.drop_duplicates(inplace=True)# deletes repeated rows and saves in original data
print(df)
print(df.reset_index(drop=True))# resets index in order
```
